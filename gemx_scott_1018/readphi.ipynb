```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Load data from files
R = np.loadtxt('R.dat')
Z = np.loadtxt('Z.dat')
testphi = np.loadtxt('./out/testphi')
masktest = np.loadtxt('./out/mask.dat')
psitest = np.loadtxt('psi_test')
netest = np.loadtxt('./out/testne')
```

```python
# Initialize arrays
phi = np.zeros((434, 450))
mask = np.zeros((434, 450))
psi = np.zeros((434, 450))
ne = np.zeros((434, 450))
```

```python
# Populate arrays
for i in range(434):
    for j in range(450):
        ii = math.floor((i*450+j)/3)
        jj = i * 450 + j - 3 * (math.floor((i*450+j)/3))
        phi[i, j] = testphi[ii, jj]
        mask[i, j] = masktest[ii, jj]
        psi[i, j] = psitest[ii, jj]
        ne[i, j] = netest[ii, jj]
```

```python
# Create a contour plot for ne
plt.contourf(R, Z, ne)
plt.axis('equal')
plt.gca().set_facecolor('white')  # Set background color
plt.colorbar()  # Add colorbar
plt.show()
plt.savefig('./out/plot_ne.png')
```

```python
# Create a contour plot for phi
plt.contourf(R, Z, phi, 100)
plt.axis('equal')
plt.gca().set_facecolor('white')  # Set background color
plt.colorbar()  # Add colorbar
plt.show()
plt.savefig('./out/plot_phi.png')
```

```python
# Create a contour plot for mask
plt.contourf(R, Z, mask)
plt.axis('equal')
plt.gca().set_facecolor('white')  # Set background color
plt.colorbar()  # Add colorbar
plt.show()
plt.savefig('./out/plot_mask.png')
```
